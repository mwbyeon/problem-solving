```python

%matplotlib inline
import matplotlib.pyplot as plt

x_axis = []
y_axis = [ [] for _ in xrange(40) ]
max_impression = {1: 2000, 2: 2000, 3: 2000, 4: 2000, 5: 2000, 6: 2000, 7: 2000, 8: 2000, 9: 2000, 10: 2000, 11: 5000, 12: 5000, 13: 5000, 14: 5000, 15: 5000, 16: 5000, 17: 5000, 18: 5000, 19: 5000, 20: 5000, 21: 10000, 22: 10000, 23: 10000, 24: 10000, 25: 10000, 26: 10000, 27: 10000, 28: 10000, 29: 10000, 30: 10000, 31: 20000, 32: 20000, 33: 20000, 34: 20000, 35: 20000, 36: 20000, 37: 20000, 38: 20000, 39: 20000, 40: 20000}
y_clicks = []
next_cpr = [ 0.0 ] * 1000
next_cpr_cnt = [ 0 ] * 1000
sum_rate = 0.0
lines = open('media_count_v2_10', 'r').readlines()
for i, line in enumerate(lines[1:]):
    line = line.strip().split()
    line2 = map(float, lines[i+1].strip().split()[4:])
    if line :
        x_axis.append(int(line[0]))
        y_clicks.append((int(line[2])))
        for y in xrange(40):
            click = int(line[y+4])
            cpr = float(click) / max_impression[y+1] / 0.974889217134
            cpr2 = line2[y] / max_impression[y+1] / 0.974889217134
            y_axis[y].append(cpr)
            p = int(cpr2*10000)
            next_cpr_cnt[p] += 1
            next_cpr[p] += (cpr2 - next_cpr[p]) / next_cpr_cnt[p]

fig, ax1 = plt.subplots(figsize=(20,10))
for i in xrange(40):
    ax1.grid(True)
    ax1.plot(x_axis, y_axis[i], '.')

fig, ax2 = plt.subplots(figsize=(20,10))
ax2.plot(next_cpr)
    
#for i, x in enumerate(next_cpr):
#    print 0.0001 * i, x
    
print sum(map(sum, y_axis))
```
